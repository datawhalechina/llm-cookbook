# 第一章 介绍 Introduction

我们设计的本课程旨在帮助开发者学会如何预处理用于 LLM 应用的非结构化数据。课程首先介绍了**检索增强生成（RAG）技术的核心组件**，包括**数据加载、分块、嵌入、存储和检索**。我们将学习如何处理不同来源的多种非结构化数据类型，例如文本、图像和表格，并将这些数据存储在向量数据库中，便于后续的检索和使用。

课程内容涵盖了**处理各种非结构化数据的技术**，帮助我们掌握处理 PDF、PowerPoint 和 Word 等不同文件格式的方法。由于数据的多样性和复杂性，课程主要聚焦**数据加载与分块方面的难题**，确保 LLM RAG 管道能够访问所需要的信息。我们还将深入探讨**多种数据格式的解析与标准化技术**，帮助处理来自 Excel 电子表格、PDF 或 Markdown 格式的文本报告、PowerPoint 或 Slides 格式的演示文稿以及 Outlook 或 Slack 等通讯工具的数据。

在数据格式化与标准化部分，我们将学习**如何解析和标准化不同来源的数据**，通过标准化，确保表格、图像、项目符号列表等数据能够以相似的方式表示。此外，课程还讲解了**元数据的使用**，通过元数据增强内容，提高 RAG 结果的准确性，维护原始文档的结构信息，例如章节标题和段落的关系。

课程还涵盖了**文档图像分析技术**，帮助我们提取和标准化内容。我们将会应用所学技术，使用 PDF、PowerPoint、Word 等文档类型**构建一个 RAGbot**。通过综合应用所学内容，我们将学习**如何将数据组织成树状结构，以实现高效查询**。
