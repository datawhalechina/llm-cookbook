## 二、为什么文档内容的规范化重要？
***

对于大语言模型（LLM）应用程序来说，文档内容的规范化至关重要，原因如下：

1. **统一处理流程**：文档规范化使得不同来源和格式的文档能够以统一的方式被处理。这样，无论文档是 PDF、Word、PowerPoint 还是 HTML 格式，LLM 都可以使用相同的逻辑和算法来解析和分析内容。

2. **提高效率**：规范化的内容可以减少处理文档时所需的重复性工作。通过预先定义好的内容结构，可以避免为每种文档类型编写单独的处理逻辑，从而提高开发效率和运行效率。

3. **增强可读性**：规范化后的内容更易于 LLM 理解和分析。通过将文档分解为标题、叙述文本等结构化元素，LLM 可以更准确地识别和提取关键信息。

4. **过滤无关信息**：规范化过程允许开发者指定哪些元素是重要的，哪些可以忽略（如页眉、页脚等）。这有助于 LLM 集中处理与任务相关的内容，提高信息的相关性和准确性。

5. **降低成本**：预处理文档的初始内容提取通常是最昂贵的部分。通过规范化，可以减少对每种文档类型特定处理的需求，从而降低整体的计算和开发成本。

6. **数据重用**：规范化后的数据可以序列化为 JSON 等格式，便于存储和重用。这意味着一旦文档被规范化，就可以在不同的应用和场景中重复使用，而无需重新预处理。

7. **支持多种操作**：规范化的内容使得可以应用统一的下游操作，如文本分块、关键词提取、内容摘要等，这对于构建多功能的 LLM 应用程序是非常有用的。

8. **提高灵活性**：规范化的内容可以轻松地适应不同的应用需求。无论是在 API 调用、Web 服务还是本地应用程序中，规范化的数据都可以无缝集成。

9. **促进创新**：当文档内容以统一格式存在时，开发者和研究人员可以更自由地尝试和实现新的算法和模型，推动 LLM 应用程序的创新和发展。

10. **跨平台兼容性**：规范化的内容可以跨不同的编程环境和平台使用，这对于构建需要在多种技术栈上运行的应用程序尤为重要。

总的来说，文档内容的规范化是构建高效、可扩展且成本效益高的 LLM 应用程序的关键步骤，它为处理和分析大量异构数据提供了坚实的基础。

如果您感觉下载速度很慢，可以手动从 Github 下载 nltk_data/packages 中的有关文件到本地，并将其重命名为 `nltk_data`。Github：https://github.com/nltk/nltk_data 。然后用以下 Python 脚本解压这个 nltk_data 目录下的所有 .zip 文件到它们各自所在的目录中。

````python
import os
import zipfile


def unzip_all_in_dir(directory):
    """
    解压指定目录下的所有 .zip 文件到它们各自所在的目录中。
    """
    # 遍历指定目录
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".zip"):
                # 构建 .zip 文件的完整路径
                zip_path = os.path.join(root, file)
                # 打开 .zip 文件
                with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                    # 解压到当前 .zip 文件所在的目录
                    zip_ref.extractall(root)
                print(f"已解压: {zip_path}")


if __name__ == "__main__":
    # 替换为您想要扫描和解压 .zip 文件的目录
    directory_to_scan = "/path/to/your/directory"
    unzip_all_in_dir(directory_to_scan)
````

🚀 **为什么要将不同文档类型的内容序列化为 JSON？**

将不同文档类型的内容序列化为 JSON 的原因包括：

1. **通用性和广为人知的结构**：JSON 是一种常见且广为人知的数据格式，这使得它易于被不同的系统和应用程序理解和使用。

2. **标准的 HTTP 响应**：在处理需要在 API 上运行模型工作负载的文档，如 PDF 和图像时，JSON 作为标准的 HTTP 响应格式，可以方便地在网络应用中传输。

3. **多语言支持**：JSON 可以在多种编程语言中使用，这意味着无论是在 JavaScript、Python 还是其他语言构建的应用程序中，都可以轻松地处理 JSON 格式的数据。

4. **流式处理用例**：JSON 也适用于流式处理用例，例如将文档存储为 JSON Lines（JSON-L），这是一种将 JSON 对象逐行存储的格式，适合日志文件或大型数据集的处理。

5. **数据重用**：数据序列化为 JSON 后，可以方便地存储和重用，无需再次预处理原始文档。这提高了效率，降低了处理成本。

6. **简化处理流程**：通过将所有文档内容规范化为 JSON 格式，可以简化后续的处理流程，如文本分析、信息抽取等，因为处理程序可以以统一的方式操作所有类型的文档。

7. **易于导航和访问**：序列化后的 JSON 对象易于导航和访问，可以方便地通过编程方式轻松提取所需的信息。

通过将文档内容序列化为 JSON，可以确保文档在预处理后以一种结构化、标准化且易于操作的格式存在，从而为 LLM 应用程序提供一致的输入，简化开发和维护过程。

使用 unstructured API 处理 PDF 文档的具体步骤如下：

1. **理解 PDF 文档的特点**：PDF 文档与 HTML 或 PowerPoint 文档不同，它们通常包含视觉线索，如文本格式设置，这些线索可以帮助识别文档中的标题、叙述性文本和表格等元素。

2. **查找标题和文本**：在 PDF 中，标题可能会被标记为粗体并加下划线，而叙述性文本可能是更长、更厚重且包含多个句子的文本，没有粗体或下划线的强调。

3. **设置 unstructured API 客户端**：因为 PDF 处理是**基于模型的工作负载**，这在计算上是昂贵的，所以使用 API 可以简化这一过程。首先，需要设置 unstructured API 客户端，这涉及到配置所需的凭据。

4. **准备 API 请求**：使用之前设置的 unstructured API 客户端来准备请求。这包括获取 PDF 文件的名称，并将其传递给 API。

5. **发出 API 请求**：准备好请求后，发出请求以处理 PDF 文件。由于这是一个基于模型的工作负载，处理过程可能需要一些时间。

6. **接收 JSON 响应**：处理完成后，API 会返回一个 JSON 格式的响应。这个响应包含了文档的元素，如标题、文本段落和表格等。

7. **探索和使用 JSON 数据**：使用 IPython 的 JSON 显示功能来探索返回的 JSON 数据。这个 JSON 数据将包含文档的规范化表示，使得可以直观地识别出标题和叙述性文本。

8. **处理和分析**：规范化后的 PDF 内容可以用于进一步的处理和分析，如文本分析、信息抽取等。

通过这个过程，用户可以有效地将 PDF 文档中的内容提取出来，并以结构化的 JSON 格式存储，便于 LLM 应用程序进一步使用和分析。

````python
# 待处理 PDF 文件的路径
filename = "example_files/CoT.pdf"

# 调用 unstructured API 的 shared.Files 类来创建一个新对象
# 此对象用于表示待处理的 PDF 文件，其中包括文件内容（通过调用 f.read() 读取）和文件名。
with open(filename, "rb") as f:
    files=shared.Files(
        content=f.read(), 
        file_name=filename,
    )

# 调用 shared.PartitionParameters 来配置如何处理上一步创建的 PDF 文件对象
# 包括使用高分辨率 ('hi_res') 策略、启用 PDF 表格结构推断 (pdf_infer_table_structure=True) 以及设置语言为英语 (languages=["eng"])。
req = shared.PartitionParameters(
    files=files,
    strategy='hi_res',
    pdf_infer_table_structure=True,
    languages=["eng"],
)
````

在一个 try-except 块中，我们尝试调用一个假定存在于 s.general.partition(req) 中的方法或函数，该函数或方法尚未定义。此次调用基于先前配置的参数（即 req）来执行实际的 PDF 处理任务。

若成功执行，程序将打印出响应对象（resp）中元素列表的前六项，并使用 JSON 格式进行美化输出。

若在此过程中抛出名为 SDKError 的异常，我们将捕获该异常并打印出相应的错误信息。这种异常处理机制确保了程序在遇到错误时不会直接崩溃，而是能够提供有用的错误反馈。

````python
try:
    resp = s.general.partition(req)
    print(json.dumps(resp.elements[:6], indent=2))
except SDKError as e:
    print(e)
````
