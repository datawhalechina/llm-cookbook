# 第二章 LLM数据预处理基础 Overview of LLM Data Preprocessing





## 一、总览
***

本章旨在帮助开发者了解LLM数据预处理的基础知识。通过本章的学习，开发者将掌握预处理各种非结构化数据的必要性，以及如何处理各种非结构化数据，以便在LLM应用中有效利用这些数据。我们将探讨检索增强生成（RAG）技术的核心组件，详细讲解文档预处理的步骤和面临的挑战，并提供实际应用的指导。通过标准化和解析不同数据格式，开发者将能够提升数据的可用性和检索效率，为构建高效的LLM应用奠定坚实的基础。

## 二、检索增强生成（RAG）技术
***

<img src="./images/2-1.png" alt="数据预处理对RAG系统的必要性" width="70%" />


理解检索增强生成（Retrieval Augmented Generation，简称RAG）技术是进行特定数据类型预处理的首要任务。RAG通过将大语言模型（LLM）的响应与经过验证的外部信息（如企业内部信息）相结合，确保模型生成的回答更加准确和具体。这些信息可以以多种文档形式存在，例如电子邮件、PDF文件或PowerPoint演示文稿。RAG应用程序通过将这些外部信息加载到数据库中，并在需要时检索这些内容，将其嵌入提示词中，再传递给LLM。这样，当LLM接收到问题或提示时，输入中已经包含了构建响应所需的外部信息。在行业应用中，这种方法尤为重要，因为组织内的数据形式多样且复杂多变。

## 三、文档预处理步骤
***

预处理文档涉及多个关键步骤。首先，必须提取文档的文本内容，这是构建提示词的基础。此外，提取文档的各类元素也是至关重要的，这些元素构成了文档的基本构建块，包括标题、叙述文本、列表和表格。通过利用这些文档元素，能够执行多种对检索增强生成（RAG）应用至关重要的任务，例如内容分块和信息过滤。在课程中，我们将会详细介绍这些操作方法。此外，提取元素的元数据（如页码和文件类型）同样具有重要意义，这些元数据在混合搜索中起着关键作用，使得在构建提示词时能够从向量数据库中有效过滤和检索信息。

## 四、数据预处理存在的挑战
***

数据预处理的复杂性源于不同文档类型具有不同的内容线索。例如，HTML文件使用标签名来指示文本是标题还是列表，而PDF文件则通过视觉线索来传达信息。为了能够统一预处理这些不同的文档类型，需要理解不同文档类型中的元素表示。此外，文档存在各种格式，需要将其标准化，以便应用程序能够以相同的方式处理它们。标准化过程使得应用程序无需关心文档是HTML还是PDF，从而可以统一处理所有文档。然而，标准化过程困难重重，因为文档格式和元素线索各异。例如，学术期刊文章与表格的处理方式不同。数据预处理技术需要理解文档结构，以提取有意义的元数据，这些元数据在RAG应用中至关重要，如过滤操作。在之后的课程中，我们将会详细探讨这些技术，帮助开发者在RAG应用中有效处理和标准化各种文档类型。
